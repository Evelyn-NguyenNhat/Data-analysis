```python
import pandas as pd

# Dataset ban đầu
data = {
    'TransactionID': range(1, 13),
    'Date': pd.to_datetime([
        '2024-01-01', '2024-01-03', '2024-01-04', '2024-02-15',
        '2024-02-17', '2024-03-01', '2024-03-15', '2024-03-20',
        '2024-04-01', '2024-04-05', '2024-04-10', '2024-04-15'
    ]),
    'CustomerID': ['C001', 'C002', 'C001', 'C003', 'C004', 'C001',
                   'C002', 'C003', 'C005', 'C002', 'C005', 'C001'],
    'Country': ['Vietnam', 'Vietnam', 'Vietnam', 'Thailand', 'Thailand', 'Vietnam',
                'Thailand', 'Thailand', 'Singapore', 'Singapore', 'Singapore', 'Vietnam'],
    'Product': ['A', 'B', 'A', 'A', 'C', 'C', 'A', 'B', 'C', 'B', 'A', 'B'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Furniture',
                 'Furniture', 'Electronics', 'Electronics', 'Furniture', 'Electronics', 'Electronics', 'Electronics'],
    'UnitPrice': [200, 150, 200, 210, 300, 310, 220, 160, 330, 170, 230, 160],
    'Quantity': [1, 2, 1, 1, 2, 1, 3, 1, 2, 2, 1, 2]
}

# Tạo DataFrame ban đầu
df = pd.DataFrame(data)
print(df)
```

```python

df = pd.DataFrame(data)
df['Revenue'] = df['UnitPrice'] * df['Quantity']
df['Month'] = df['Date'].dt.to_period('M')
df.head()
```

```python
#1 Tổng doanh thu theo quốc gia
df.groupby('Country')['Revenue'].sum()
```

```python
# Trung bình doanh thu mỗi giao dịch theo sản phẩm
```

```python
df.groupby('Product')['Revenue'].mean()
```

```python
# Số lượng giao dịch theo quốc gia
```

```python
df.groupby('Country')['TransactionID'].count()
```

```python
#Tổng số lượng bán theo sản phẩm
df.groupby('Product')['Quantity'].sum()
```

```python
#PHẦN 3: SỬ DỤNG .AGG() ĐỂ KẾT HỢP NHIỀU HÀM
```

```python
df.groupby('Product').agg({
    'Revenue': ['sum','mean'],
    'Quantity': 'sum'
})
```

```python
#PHẦN 4: LÀM VIỆC VỚI NHIỀU CỘT NHÓM
```

5. Tổng doanh thu theo quốc gia và sản phẩm

```python
df.groupby(['Country','Product'])['Revenue'].sum()
```

```python
# Tạo bảng dạng pivot để dễ đọc
```

```python
df.groupby(['Country', 'Product'])['Revenue'].sum().unstack()
```

```python
# Loc các khách hàng có tổng doanh thu > 600
df.groupby('CustomerID').filter(lambda x: x['Revenue'].sum() > 600)
```

Ví dụ 2: Lọc các sản phẩm có trung bình số lượng bán ra > 1.5 mỗi giao dịch

```python
df.groupby('Product').filter(lambda x: x['Quantity'].mean() > 1.5)
```

```python
df['RevvenueInGroup']= df.groupby('Country')['Revenue'].transform('sum')
print(df)
```

```python
# VD4: TÌM CÁC QUỐC GIA CÓ ÍT NHẤT 3 GIAO DỊCH
```

```python
df.groupby('Country').filter(lambda x: len(x)>=3)
```

```python
#Lọc ra các khách hàng có trung bình giá trị mỗi đơn hàng (Revenue) > 250.
```

```python
df.groupby('CustomerID').filter(lambda x: x['Revenue'].mean()>250)
```

```python
df['AvgPriceByCategory']=df.groupby('Category')['UnitPrice'].transform('mean')
print(df)
```

```python
#Buoc 1
high_revenue_countries=df.groupby('Country')['Revenue'].sum().loc[lambda x:x>1000].index
#Buoc 2
filtered_df=df[df['Country'].isin(high_revenue_countries)]
#Buoc 3
result=filtered_df.groupby(['Country','Product'])['Quantity'].sum().reset_index()
print(result)
```

```python
#B1: Lấy các loại Category mà mỗi khách đã mua
customer_categories= df.groupby('CustomerID')['Category'].unique()
#B2: Tìm các khách hàng không có 'Furniture' trong danh sách
electronics_only_customers=customer_categories[~customer_categories.apply(lambda x:
'Furniture' in x)].index
#B3: Lọc lại Dataframe gốc để lấy giao dịch của các khách hàng này
result_df=df[df['CustomerID'].isin(electronics_only_customers)]
print(result_df)
```
